### Una empresa eléctrica necesita un programa que permita monitorear la corriente que circula por una línea de distribución.

```python
while True:
    corriente_nominal = float(input("ingresar corriente nominal de la linea (A) "))
    while corriente_nominal < 0:
        print("valor invalido intente de nuevo")
        corriente_nominal = float(input("ingresar corriente nominal de la linea (A) "))

    corriente_media = float(input("ingrese la corriente media (A)"))
    while corriente_media <= 0:
        print("La corriente media no puede ser cero o negativa. Intente de nuevo.")
        corriente_media = float(input("ingrese la corriente media (A)"))

    porcentaje = float((corriente_nominal / corriente_media) * 100)
    print(f"el porcentaje de carga es: {porcentaje}%")

    if porcentaje <= 70:
        print("operacion normal")
    elif porcentaje > 70 and porcentaje <= 90:
        print("carga alta")
    elif porcentaje > 90 and porcentaje <= 100:
        print("alerta")
    else:
        print("sobrecarga, Riesgo de disparo")
    repetir = str(input("desea repetir la operacion? (s/n)"))
    if repetir == "n":
        break
```
